```python
import cv2
img = cv2.imread('/content/WhatsApp Image 2024-05-25 at 15.08.18_6e695b99.jpg')

#Reading an Image:
#Load an image from a file.
```

```python
#Writing an Image:
#Save an image to a file.
cv2.imwrite('/content/WhatsApp Image 2024-05-25 at 15.08.18_6e695b99.jpg', img)
```

```python
cv2.circle(img, (250, 150), 50, (255, 0, 0), -1)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
import matplotlib.pyplot as plt

edges = cv2.Canny(img, 100, 200)

plt.imshow(edges, cmap='Reds')
plt.show()
```

```python
#cap = cv2.VideoCapture(0)
```

```python
resized = cv2.resize(img, (200, 200))
cv2.imwrite('/content/output.jpg', resized)
```

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
cv2.imwrite('/content/output3.jpg', img)
```
